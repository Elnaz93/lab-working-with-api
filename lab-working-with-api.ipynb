```python
# Create a function to automate the work done in Step 1
def get_audio_features(artist):
    # get tracks from artist
    results = sp.search(q=f'artist:{artist}', limit=50)
    # extract the track ids
    try: # Some artists have no items (e.g., collabs)
        track_ids = [track['id'] for track in results['tracks']['items']]
        song_names = [track['name'] for track in results['tracks']['items']]
        # extract the audio features
        audio_features = sp.audio_features(track_ids)
        # store audio features in a dataframe
        df = pd.DataFrame(audio_features)
        df['artist'] = artist
        df['song name'] = song_names
        return df
    except: # If an artist has no items, function returns nothing
        return None
```

```python
# Create loop and select artists you want to get songs from

artists = ['Windows 96', 'bbrainz', 'Home']

df = pd.DataFrame()

for artist in artists:
    df_artist = get_audio_features(artist)
    df = pd.concat([df, df_artist])
    
df = df.set_index("id").drop(["type", "uri", "track_href", "analysis_url"], axis=1)
df = df[['artist', 'song name', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_ms', 'time_signature']]
df
```

```python
# Create a function to get all the tracks from a playlist

def get_playlist_tracks(username, playlist_id):
    
    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']
    
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    
    return tracks
```

```python
# Create a function to get all the artist names featuring in a playlist

def get_artists_from_playlist(username, playlist_id):
    
    tracks_from_playlist = get_playlist_tracks(username, playlist_id)
    
    artists = []
    
    for track in tracks_from_playlist:
        artists_info = track['track']['artists']
        
        for artist_info in artists_info:
            artists.append(artist_info['name']) #change for ['id'] to get IDs
    
    return list(set(artists))
```

```python
# Create a loop to get songs from all the artists featuring in a playlist

artists = get_artists_from_playlist(username="spotify", playlist_id="37i9dQZF1DXcBWIGoYBM5M")

df = pd.DataFrame()

for artist in artists:
    df_artist = get_audio_features(artist)
    df = pd.concat([df, df_artist])
    
df = df.set_index("id").drop(["type", "uri", "track_href", "analysis_url"], axis=1)
df = df[['artist', 'song name', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_ms', 'time_signature']]
df
```
